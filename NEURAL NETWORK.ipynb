### CODE FOR NEURAL NETWORK

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
### DATA
df=pd.read_csv("C:/Users/Jose de Vera/Desktop/AI 201/miniproject/standardized.csv")
input = df.iloc[:,0:4]
output = df.iloc[:,4:]

### SPLIT
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size=0.3, random_state=201)
```

```python
y_train.to_csv("C:/Users/Jose de Vera/Desktop/AI 201/miniproject/train_outputs.csv",index=False)
y_test.to_csv("C:/Users/Jose de Vera/Desktop/AI 201/miniproject/test_outputs.csv",index=False)
```

```python
max_epochs = 300
nodes = 200
# Build the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(nodes, activation='linear', input_shape=(4,)),
    tf.keras.layers.Dense(200, activation='linear', input_shape=(nodes,)),
    tf.keras.layers.Dense(5)
])

# Compile the model
opt = tf.keras.optimizers.Adam(learning_rate=0.05)
model.compile(optimizer=opt, loss='mse', metrics=['mae'])

# Lists to store training and test MSE values
train_mse_values = []
test_mse_values = []

# MSE over number of epochs
for epoch in range(max_epochs):
    print("Currently on epoch: "+str(epoch))
    history = model.fit(x_train, y_train, epochs=1, batch_size=15, validation_data=(x_test, y_test), verbose=0)
    
    # Record training and test MSE
    train_mse = history.history['loss'][0]
    test_mse = history.history['val_loss'][0]
    
    train_mse_values.append(train_mse)
    test_mse_values.append(test_mse)
```

```python
nodes = 75
# Build the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(nodes, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(50, activation='relu', input_shape=(nodes,)),
    tf.keras.layers.Dense(5)
])
# Compile the model
opt = tf.keras.optimizers.Adam(learning_rate=0.05)
model.compile(optimizer=opt, loss='mse', metrics=['mae'])

model.fit(x_train, y_train, epochs=230, batch_size=20, validation_data=(x_test, y_test), verbose=0)
```

```python
y_train_pred = model.predict(x_train)
y_train_pred = pd.DataFrame(y_train_pred, columns=['ThalPredict', 'ChaePredict', 'BactPredict','PseuPredict','CoscPredict'])
y_test_pred = model.predict(x_test)
y_test_pred = pd.DataFrame(y_test_pred, columns=['ThalPredict', 'ChaePredict', 'BactPredict','PseuPredict','CoscPredict'])
y_test_pred
```
