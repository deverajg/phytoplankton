### CODE FOR PERCEPTRON ENSEMBLE

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
### DATA
df=pd.read_csv("C:/Users/Jose de Vera/Desktop/AI 201/miniproject/standardized.csv")
input = df.iloc[:,0:4]
output = df.iloc[:,4:]

### SPLIT
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size=0.3, random_state=201)
```

```python
# Parameters for the ensemble
n_learners = 50
learners = []
train_mse_values = []
test_mse_values = []

# Train the ensemble
for i in range(1, n_learners + 1):
    # Create a single-layer perceptron as the weak learner
    base_regressor = MLPRegressor(hidden_layer_sizes=(10,), max_iter=1000, random_state=201)
    
    # Create a bagging regressor with the weak learner
    bagging_regressor = BaggingRegressor(base_estimator=base_regressor, n_estimators=i, random_state=201)

    # Train the ensemble on the training data
    bagging_regressor.fit(x_train, y_train)

    # Make predictions on the test data
    train_pred = bagging_regressor.predict(x_train)
    test_pred = bagging_regressor.predict(x_test)

    # Calculate and store the mean squared error
    train_mse = mean_squared_error(y_train,train_pred)
    train_mse_values.append(train_mse)

    test_mse = mean_squared_error(y_test, test_pred)
    test_mse_values.append(test_mse)

    print(f'Number of learners: {i}, MSE: {test_mse}')

    # Save the trained learner for plotting later
    learners.append(bagging_regressor)


```

```python
# Parameters for the ensemble
n_learners = 50
learners = []
train_mse_values1 = []
test_mse_values1 = []

# Train the ensemble
for i in range(1, n_learners + 1):
    # Create a single-layer perceptron as the weak learner
    base_regressor = DecisionTreeRegressor(max_depth=5)
    
    # Create a bagging regressor with the weak learner
    bagging_regressor = BaggingRegressor(base_estimator=base_regressor, n_estimators=i, random_state=201)

    # Train the ensemble on the training data
    bagging_regressor.fit(x_train, y_train)

    # Make predictions on the test data
    train_pred = bagging_regressor.predict(x_train)
    test_pred = bagging_regressor.predict(x_test)

    # Calculate and store the mean squared error
    train_mse = mean_squared_error(y_train,train_pred)
    train_mse_values1.append(train_mse)

    test_mse = mean_squared_error(y_test, test_pred)
    test_mse_values1.append(test_mse)

    print(f'Number of learners: {i}, MSE: {test_mse}')

    # Save the trained learner for plotting later
    learners.append(bagging_regressor)


```

```python
# Parameters for the ensemble
n_learners = 50
learners = []
train_mse_values2 = []
test_mse_values2 = []

# Train the ensemble
for i in range(1, n_learners + 1):
    # Create a single-layer perceptron as the weak learner
    base_regressor = KNeighborsRegressor(n_neighbors=5)
    
    # Create a bagging regressor with the weak learner
    bagging_regressor = BaggingRegressor(base_estimator=base_regressor, n_estimators=i, random_state=201)

    # Train the ensemble on the training data
    bagging_regressor.fit(x_train, y_train)

    # Make predictions on the test data
    train_pred = bagging_regressor.predict(x_train)
    test_pred = bagging_regressor.predict(x_test)

    # Calculate and store the mean squared error
    train_mse = mean_squared_error(y_train,train_pred)
    train_mse_values2.append(train_mse)

    test_mse = mean_squared_error(y_test, test_pred)
    test_mse_values2.append(test_mse)

    print(f'Number of learners: {i}, MSE: {test_mse}')

    # Save the trained learner for plotting later
    learners.append(bagging_regressor)


```

```python
plt.plot(range(1, n_learners + 1), train_mse_values, marker='o',label = "MLP regressor ensemble")
plt.plot(range(1, n_learners + 1), train_mse_values2, marker='o', label = "KNN regressor ensemble")
plt.plot(range(1, n_learners + 1), train_mse_values1, marker='o',label = "Decision tree ensemble")
plt.title('Average MSE vs. Number of Learners')
plt.xlabel('Number of Learners')
plt.ylabel('Average Mean Squared Error')
plt.legend()
plt.show()
```

```python
plt.plot(range(1, n_learners + 1), test_mse_values, marker='o',label = "MLP regressor ensemble")
plt.plot(range(1, n_learners + 1), test_mse_values2, marker='o', label = "KNN regressor ensemble")
plt.plot(range(1, n_learners + 1), test_mse_values1, marker='o',label = "Decision tree ensemble")
plt.title('Average MSE vs. Number of Learners')
plt.xlabel('Number of Learners')
plt.ylabel('Average Mean Squared Error')
plt.legend()
plt.show()
```
